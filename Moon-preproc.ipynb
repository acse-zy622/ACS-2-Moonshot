```python
import pandas as pd
import numpy as np
from PIL import Image
from numpy import asarray
import matplotlib.patches as patches
import matplotlib.pyplot as plt
 
train = pd.read_csv('A/A_0020_filtered.csv', sep=' ')
imgarr = asarray(Image.open('A/A_0020.jpg'))

x = (train.iloc[:,2]*imgarr.shape[1]).tolist()
y = (train.iloc[:,1]*imgarr.shape[0]).tolist()
w = (train.iloc[:,3]*imgarr.shape[0]).tolist()
h = w

fig, ax = plt.subplots() 
ax.imshow(imgarr)
ax.plot(x, y, 'r.', markersize=2)
for i in range(len(train.index)):
    rect = patches.Rectangle((x[i]-h[i]/2, y[i]-w[i]/2), h[i], w[i], linewidth=1, edgecolor='r', facecolor='none')
    ax.add_patch(rect)
plt.show()
```
